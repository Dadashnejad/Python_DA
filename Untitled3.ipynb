Develop a Bubble plot for displaying the impact of seasonality on
Automobile Sales.[link text](https://)

```python
def month_season(x):
  if x in ["Mar", "Apr", "May"]:
    return "Spring"
  elif x in ["Jun", "Jul", "Aug"]:
    return "Summer"
  elif x in ["Sep", "Oct", "Nov"]:
    return "Fall"
  elif x in ["Dec", "Jan", "Feb"]:
    return "Winter"
  else:
    return "Invalid month"
df["Season"]=df["Month"].apply(month_season)

df["YearSeason"] = df.Year.astype(str)+" "+ df.Season

df["Recession_Season"]=df.groupby("YearSeason")["Recession"].transform("mean").ge(0.6).astype(bool)
df.head()

```

```python
bubble_data=df.groupby(by="YearSeason").agg(Sales=("Price","sum"),Weight=("Seasonality_Weight","mean"),Recession=("Recession_Season","last")).reset_index(drop=False)
bubble_data.head()
```

```python
plt.figure(figsize=(40,10))
sns.scatterplot(data=bubble_data,x="YearSeason",y="Sales",size="Weight",hue="Recession",sizes=(100,300),alpha=0.8)
plt.xticks(rotation=90)
plt.show()
```

Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods

```python
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

recession_ad_expenditure = recession_data['Advertising_Expenditure'].sum()
non_recession_ad_expenditure = non_recession_data['Advertising_Expenditure'].sum()

labels = ['Recession', 'Non-Recession']
sizes = [recession_ad_expenditure, non_recession_ad_expenditure]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=(0, 0.07) , startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Advertising Expenditure During Recession and Non-Recession Periods')
plt.axis('equal')
plt.show()
```

Develop a pie chart to display the total Advertisement expenditure for each vehicle type during recession period

```python
ad_expenditure_by_vehicle_type = recession_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

labels = ad_expenditure_by_vehicle_type.index
sizes = ad_expenditure_by_vehicle_type.values

plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=40)
plt.title('Advertisement Expenditure')

plt.show()
```
